```python
import pandas as pd

def load_data():
    url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
    html = pd.read_html(url, header = 0)
    df = html[0]
    return df
```

```python
df = load_data()
df
```

```python
#  Setores de empresas
setor_unico = df['GICS Sector'].unique()
setor_unico
```

```python
#  Quantidade de setores de empresas
len(setor_unico)
```

```python
#  Empresas por setor
setor = df.groupby('GICS Sector')
setor.describe()
```
